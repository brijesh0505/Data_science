```python
import seaborn as sns
```

```python
df=sns.load_dataset('iris')
```

```python
df.head()
```

```python
# define corelation with data
# it must be in int or float not in object
df.corr()
```

```python
df.info()
```

```python
sns.heatmap(df.corr())
```

```python
# histogram
sns.jointplot(x='sepal_length',y='sepal_width',data=df,kind='hex')
```

```python
# regration
sns.jointplot(x='sepal_length',y='sepal_width',data=df,kind='reg')
```

# pair plot

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='species')
```

```python
df['species'].value_counts()
```

# Categorial plot

```python
df=sns.load_dataset('tips')
```

```python
df
```

```python
df.tail()
```

```python
sns.countplot(y='day',data=df)
```

```python
sns.barplot(x='sex',y='total_bill',data=df)
```

```python
sns.boxplot('sex','total_bill',data=df)
```

```python
sns.boxplot(x='day',y='total_bill',data=df)
```

```python
sns.boxplot(x='total_bill',y='day',hue='smoker',data=df)
```

# violin plot

```python
sns.violinplot(x='total_bill',y='day',data=df)
```
